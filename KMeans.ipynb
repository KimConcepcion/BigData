```python
## Import stuff:
from sklearn import cluster
import pandas as pds

## matplotlib:
%pylab inline
```

```python
## Read the csv file:
data = pds.read_csv('/Users/kimnielsen/JupyterNotebook/BigData/smartphone.csv')

## The type of the data, ensure that it is a panda frame:
type(data)

## List the data:
data.index = data['loggingTime']
data
```

```python
## The values of interest:
acc = data[['accelerometerAccelerationX', 'accelerometerAccelerationY', 'accelerometerAccelerationZ']]
gyro = data[['gyroRotationX', 'gyroRotationY', 'gyroRotationZ']]
```

```python
## Initialize model with 6 clusters:
model = cluster.KMeans(n_clusters=6)

## Model:
model
```

```python
acc_model = model.fit(acc)
gyro_model = model.fit(gyro)
```

```python
## X, Y and Z with 6 clusters:
print('Acceleration:')
acc_model.cluster_centers_
```

```python
print('Gyroscope rotation:')
gyro_model.cluster_centers_
```

```python
## Prediction of acceleration - convert acc numbers into cluster numbers - They become numpy arrays:
prec_acc = model.predict(acc)
prec_gyro = model.predict(gyro)

## Total size of all clusters:
print('Size of cluster for acceleration:', len(prec_acc))
print('Size of cluster for gyroscope rotation:', len(prec_gyro))

## List the prediction:
prec_acc
prec_gyro

## Divide the clusters in 3:
clusters_lenghts_acc = [len(prec_acc[prec_acc==x]) for x in range(6)]
print('Acc:  ', clusters_lenghts_acc)

clusters_lenghts_gyro = [len(prec_gyro[prec_gyro==x]) for x in range(6)]
print('Gyro: ', clusters_lenghts_gyro)
```
