```python
%pylab inline

from sklearn.svm import SVR
import sklearn.datasets as ds
import random
from sklearn import svm
```

```python
iris = ds.load_iris()

## Classes:
print(iris.target_names)
print(iris.target)
```

```python
figure(), plot(iris.target), grid()
```

```python
##model = svm.SVC(kernel='linear')
##model = svm.LinearSVR()
model = svm.SVC(kernel='linear')
```

```python
## Supervised, since SVM needs a bit info before it works:
irLen = len(iris.target)
train_vector = [random.randint(0, irLen-1) for i in range(irLen)]
train_data = iris.data[train_vector]
target_data = iris.target[train_vector]
```

```python
model_train = model.fit(train_data, target_data)
model_prec = model.predict(iris.data)
```

```python
plot(model_prec)
```

```python
figure, ax = plt.subplots(figsize=(15, 10))
ax.plot(iris.target, 'r'), grid(), plot(model_prec, 'g'), title('Prediction vs. data')
```
