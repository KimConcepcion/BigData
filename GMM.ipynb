```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
root = '/Users/kimnielsen/JupyterNotebook/Exercises/'
data = pd.read_csv(root + 'smartphone.csv')
```

```python
data
```

```python
data.index = data['loggingTime']
```

```python
data
```

```python
type(data)
```

```python
acc = data[['accelerometerAccelerationX', 'accelerometerAccelerationY', 'accelerometerAccelerationZ']]
```

```python
type(acc)
```

```python
n_clusters = 0
model = KMeans(n_clusters=6)
```

```python
model.fit(acc)
```

```python
model.cluster_centers_
```

```python
res = model.predict(acc)
```

```python
res
```

```python
cluster_size = len(res)
```

```python
cluster_size
```

```python
cluster_size = [len(res[res==x]) for x in range(6)]
```

```python
cluster_size
```

```python
## Gaussian Mixture Model
from sklearn.mixture import GaussianMixture
```

```python
model = GaussianMixture(n_components=6)
```

```python
model = model.fit(acc)
```

```python
model
```

```python
res2 = model.predict(acc)
```

```python
data['GMM'] = res2
```
