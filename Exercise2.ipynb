```python
## Import libraries:
from numpy import sinc

## This includes functions like linspace etc.
%pylab inline
x = linspace(-4, 4)
y_sinc_1 = sinc(x)
y_sinc_2 = sin(x)/x

## Figure and plot handling:
figure()
plot(y_sinc_1, 'r', label='y_sinc_1')
title('Plot af sinc funktion - Numpy'), ylabel('y_sinc_1(x)'), xlabel('x'), grid(), legend(loc='best')

figure()
plot(y_sinc_2, 'c', label='y_sinc_2')
title('Plot af sinc funktion - Triviel'), ylabel('y_sinc_2(x)'), xlabel('x'), grid(), legend(loc='best')
```
